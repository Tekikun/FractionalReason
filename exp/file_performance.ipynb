```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------------ user config ------------------
path = "/home/woodywu/Desktop/Research/Steering/FractionalReason/records/plain_majority_vote"

# file lists (adjust or auto-discover if you want)
cot_files = [
    'Qwen2.5_woodywu-APBench_0_100_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'Qwen2.5_woodywu-APBench_0_100_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'Qwen2.5_woodywu-APBench_0_100_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'Qwen2.5_woodywu-APBench_0_100_ntrials_7_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote'
]

classifier_files = [
    'Qwen2.5_woodywu-APBench_0_80_icv_classifier_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_classifier_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_classifier_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]

knowledge_files = [
    'Qwen2.5_woodywu-APBench_0_80_icv_knowledge_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_knowledge_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_knowledge_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]

math_files = [
    'Qwen2.5_woodywu-APBench_0_80_icv_math_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_math_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_math_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]

groups = {
    'CoT': cot_files,
    'classifier': classifier_files,
    'knowledge': knowledge_files,
    'math': math_files
}

# color per icv type (change to any colors you like)
color_map = {
    'CoT': 'C0',
    'classifier': 'C1',
    'knowledge': 'C2',
    'math': 'C3'
}

# whether to start plotting from 10th data point (index 9)
start_from_10th = True
start_idx = 9  # zero-indexed

# how many rows to keep from each file
keep_n = 80
# -------------------------------------------------

# storage: for each group -> list of Series indexed by query_id
group_series = {g: [] for g in groups.keys()}

for icv_type, file_list in groups.items():
    for fname in file_list:
        full_path = os.path.join(path, f"{fname}.json")
        if not os.path.exists(full_path):
            print(f"Warning: file not found: {full_path}  skipping")
            continue

        # read NDJSON, keep first keep_n rows
        df = pd.read_json(full_path, lines=True).head(keep_n)

        # optionally start from 10th data point onward
        if start_from_10th:
            df = df.iloc[start_idx:].reset_index(drop=True)

        # ensure columns exist
        if 'query_id' not in df.columns or 'acc_so_far' not in df.columns:
            print(f"Warning: missing columns in {fname}, skipping")
            continue

        # build a Series indexed by query_id
        s = pd.Series(df['acc_so_far'].values, index=df['query_id'].values, name=fname)
        # convert index to integers
        s.index = s.index.astype(int)
        group_series[icv_type].append(s)

# --- assume group_series is a dict: icv_type -> list of Series (each Series.name == fname) ---

# Prepare plotting
plt.figure(figsize=(6, 4))
ax = plt.gca()

legend_handles = []
legend_labels = []

for icv_type, series_list in group_series.items():
    if len(series_list) == 0:
        continue

    # --- collapse duplicates in each series and ensure unique column names ---
    cleaned = []
    for s in series_list:
        # if s has duplicate query_id indices, aggregate them (mean)
        if s.index.duplicated().any():
            s = s.groupby(level=0).mean()
        # ensure name is unique (use filename or add suffix if needed)
        col_name = s.name
        # If somehow duplicate column names exist across files, make unique
        # by appending a counter (rare if file names are unique)
        # but pandas.concat with keys will also handle this; here we rename to filename
        s = s.rename(col_name)
        cleaned.append(s)

    # form DataFrame by aligning on the union of query_id indices
    # use keys so columns are the original filenames (guaranteed unique)
    df_concat = pd.concat(cleaned, axis=1)

    # sort by query_id (index)
    df_concat = df_concat.sort_index()

    # compute mean and std across columns (skip NaNs)
    mean_series = df_concat.mean(axis=1, skipna=True)
    std_series  = df_concat.std(axis=1, ddof=0, skipna=True)  # ddof=0 or ddof=1 as you prefer

    # x-values (query ids)
    x = mean_series.index.values
    y = mean_series.values
    yerr = std_series.values

    color = color_map.get(icv_type, None)

    # plot mean line
    line, = ax.plot(x, y, label=icv_type, color=color, linewidth=2)

    # shaded std area (mean +/- std)
    ax.fill_between(x, y - yerr, y + yerr, alpha=0.2, color=color)

    # store for legend (one handle per icv_type)
    legend_handles.append(line)
    legend_labels.append(icv_type)

# Finalize plot
ax.set_xlabel('Query ID')
ax.set_ylabel('Accuracy')
# ax.set_title(f'Accuracy evolution (first {keep_n} samples, start at 10th data point)')
ax.grid(True)

# legend on top with one row per icv_type
ax.legend(legend_handles, legend_labels, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=len(legend_handles), frameon=False)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('accuracy_evolution_mean_std_by_icvtype_Qwen2.5.pdf', bbox_inches='tight', format='pdf')
plt.show()

```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------------ user config ------------------
path = "/home/woodywu/Desktop/Research/Steering/FractionalReason/records/plain_majority_vote"

# file lists (adjust or auto-discover if you want)
cot_files = [
    'llama-3_woodywu-APBench_0_50_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'llama-3_woodywu-APBench_0_50_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'llama-3_woodywu-APBench_0_50_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'llama-3_woodywu-APBench_0_50_ntrials_7_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote'
]

classifier_files = [
    'llama-3_woodywu-APBench_0_80_icv_classifier_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_classifier_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_classifier_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]

knowledge_files = [
    'llama-3_woodywu-APBench_0_80_icv_knowledge_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_knowledge_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_knowledge_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]

math_files = [
    'llama-3_woodywu-APBench_0_80_icv_math_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_math_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_math_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]

groups = {
    'CoT': cot_files,
    'classifier': classifier_files,
    'knowledge': knowledge_files,
    'math': math_files
}

# color per icv type (change to any colors you like)
color_map = {
    'CoT': 'C0',
    'classifier': 'C1',
    'knowledge': 'C2',
    'math': 'C3'
}

# whether to start plotting from 10th data point (index 9)
start_from_10th = True
start_idx = 9  # zero-indexed

# how many rows to keep from each file
keep_n = 80
# -------------------------------------------------

# storage: for each group -> list of Series indexed by query_id
group_series = {g: [] for g in groups.keys()}

for icv_type, file_list in groups.items():
    for fname in file_list:
        full_path = os.path.join(path, f"{fname}.json")
        if not os.path.exists(full_path):
            print(f"Warning: file not found: {full_path}  skipping")
            continue

        # read NDJSON, keep first keep_n rows
        df = pd.read_json(full_path, lines=True).head(keep_n)

        # optionally start from 10th data point onward
        if start_from_10th:
            df = df.iloc[start_idx:].reset_index(drop=True)

        # ensure columns exist
        if 'query_id' not in df.columns or 'acc_so_far' not in df.columns:
            print(f"Warning: missing columns in {fname}, skipping")
            continue

        # build a Series indexed by query_id
        s = pd.Series(df['acc_so_far'].values, index=df['query_id'].values, name=fname)
        # convert index to integers
        s.index = s.index.astype(int)
        group_series[icv_type].append(s)

# --- assume group_series is a dict: icv_type -> list of Series (each Series.name == fname) ---

# Prepare plotting
plt.figure(figsize=(6, 4))
ax = plt.gca()

legend_handles = []
legend_labels = []

for icv_type, series_list in group_series.items():
    if len(series_list) == 0:
        continue

    # --- collapse duplicates in each series and ensure unique column names ---
    cleaned = []
    for s in series_list:
        # if s has duplicate query_id indices, aggregate them (mean)
        if s.index.duplicated().any():
            s = s.groupby(level=0).mean()
        # ensure name is unique (use filename or add suffix if needed)
        col_name = s.name
        # If somehow duplicate column names exist across files, make unique
        # by appending a counter (rare if file names are unique)
        # but pandas.concat with keys will also handle this; here we rename to filename
        s = s.rename(col_name)
        cleaned.append(s)

    # form DataFrame by aligning on the union of query_id indices
    # use keys so columns are the original filenames (guaranteed unique)
    df_concat = pd.concat(cleaned, axis=1)

    # sort by query_id (index)
    df_concat = df_concat.sort_index()

    # compute mean and std across columns (skip NaNs)
    mean_series = df_concat.mean(axis=1, skipna=True)
    std_series  = df_concat.std(axis=1, ddof=0, skipna=True)  # ddof=0 or ddof=1 as you prefer

    # x-values (query ids)
    x = mean_series.index.values
    y = mean_series.values
    yerr = std_series.values

    color = color_map.get(icv_type, None)

    # plot mean line
    line, = ax.plot(x, y, label=icv_type, color=color, linewidth=2)

    # shaded std area (mean +/- std)
    ax.fill_between(x, y - yerr, y + yerr, alpha=0.2, color=color)

    # store for legend (one handle per icv_type)
    legend_handles.append(line)
    legend_labels.append(icv_type)

# Finalize plot
ax.set_xlabel('Query ID')
ax.set_ylabel('Accuracy')
# ax.set_title(f'Accuracy evolution (first {keep_n} samples, start at 10th data point)')
ax.grid(True)

# legend on top with one row per icv_type
ax.legend(legend_handles, legend_labels, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=len(legend_handles), frameon=False)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('accuracy_evolution_mean_std_by_icvtype_llama3.pdf', bbox_inches='tight', format='pdf')
plt.show()

```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------- user config -------------
path = "/home/woodywu/Desktop/Research/Steering/FractionalReason/records/plain_majority_vote"

# llama-3 files
llama_cot_files = [
    'llama-3_woodywu-APBench_0_50_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'llama-3_woodywu-APBench_0_50_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'llama-3_woodywu-APBench_0_50_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'llama-3_woodywu-APBench_0_50_ntrials_7_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote'
]
llama_classifier_files = [
    'llama-3_woodywu-APBench_0_80_icv_classifier_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_classifier_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_classifier_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]
llama_knowledge_files = [
    'llama-3_woodywu-APBench_0_80_icv_knowledge_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_knowledge_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_knowledge_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]
llama_math_files = [
    'llama-3_woodywu-APBench_0_80_icv_math_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_math_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'llama-3_woodywu-APBench_0_80_icv_math_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]

# qwen2.5 files
qwen_cot_files = [
    'Qwen2.5_woodywu-APBench_0_100_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'Qwen2.5_woodywu-APBench_0_100_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'Qwen2.5_woodywu-APBench_0_100_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote',
    'Qwen2.5_woodywu-APBench_0_100_ntrials_7_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FR_majority_vote'
]
qwen_classifier_files = [
    'Qwen2.5_woodywu-APBench_0_80_icv_classifier_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_classifier_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_classifier_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]
qwen_knowledge_files = [
    'Qwen2.5_woodywu-APBench_0_80_icv_knowledge_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_knowledge_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_knowledge_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]
qwen_math_files = [
    'Qwen2.5_woodywu-APBench_0_80_icv_math_ntrials_1_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_math_ntrials_3_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote',
    'Qwen2.5_woodywu-APBench_0_80_icv_math_ntrials_5_alpha_mode_uniform_alpha_a_0_alpha_b_0.15_FA_majority_vote'
]

# grouping: keys are (model_name, icv_type) -> file list
groups = {
    ('llama-3', 'CoT'): llama_cot_files,
    ('llama-3', 'classifier'): llama_classifier_files,
    ('llama-3', 'knowledge'): llama_knowledge_files,
    ('llama-3', 'math'): llama_math_files,
    ('qwen2.5', 'CoT'): qwen_cot_files,
    ('qwen2.5', 'classifier'): qwen_classifier_files,
    ('qwen2.5', 'knowledge'): qwen_knowledge_files,
    ('qwen2.5', 'math'): qwen_math_files
}

# color per icv type (same color across models)
color_map = {
    'CoT': 'C0',
    'classifier': 'C1',
    'knowledge': 'C2',
    'math': 'C3'
}

# linestyle per model
linestyle_map = {
    'llama-3': '-',     # solid
    'qwen2.5': '--'     # dashed
}

# params
keep_n = 80            # keep first 80 rows
start_idx = 9          # start from 10th data point (index 9)

# collect per-group series
group_series = {}  # (model,icv) -> list of Series (indexed by query_id)
for key in groups.keys():
    group_series[key] = []

for (model, icv), file_list in groups.items():
    for fname in file_list:
        full_path = os.path.join(path, f"{fname}.json")
        if not os.path.exists(full_path):
            print(f"Warning: missing {full_path}, skipping.")
            continue

        # read NDJSON and trim
        df = pd.read_json(full_path, lines=True).head(keep_n)
        if start_idx is not None:
            df = df.iloc[start_idx:].reset_index(drop=True)

        # sanity check
        if 'query_id' not in df.columns or 'acc_so_far' not in df.columns:
            print(f"Warning: file {fname} missing columns, skipping.")
            continue

        s = pd.Series(df['acc_so_far'].values, index=df['query_id'].values, name=fname)
        s.index = s.index.astype(int)

        # collapse duplicate indices by mean if any
        if s.index.duplicated().any():
            s = s.groupby(level=0).mean()

        group_series[(model, icv)].append(s)

# Plot mean +/- std for each (model, icv)
plt.figure(figsize=(9, 6))
ax = plt.gca()

legend_handles = []
legend_labels = []

for (model, icv), series_list in group_series.items():
    if len(series_list) == 0:
        continue

    # concat aligned by query_id (columns = filenames)
    df_concat = pd.concat(series_list, axis=1)   # columns are filenames (unique)
    df_concat = df_concat.sort_index()

    mean_series = df_concat.mean(axis=1, skipna=True)
    std_series  = df_concat.std(axis=1, ddof=0, skipna=True)

    x = mean_series.index.values
    y = mean_series.values
    yerr = std_series.values

    color = color_map.get(icv, None)
    ls = linestyle_map.get(model, '-')
    label = f"{model} {icv}"

    # draw mean
    line, = ax.plot(x, y, linestyle=ls, color=color, linewidth=2, label=label)
    # draw shaded std
    ax.fill_between(x, y - yerr, y + yerr, color=color, alpha=0.18)

    legend_handles.append(line)
    legend_labels.append(label)

# formatting
ax.set_xlabel('Query ID')
ax.set_ylabel('Accuracy')
ax.grid(True)

# legend on top in two rows if many entries
ncols = 4
ax.legend(legend_handles, legend_labels, loc='upper center', bbox_to_anchor=(0.5, 1.12),
          ncol=ncols, frameon=False)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('combined_icv_mean_std_llama_qwen.pdf', bbox_inches='tight', format='pdf')
plt.show()
```
